```python
import numpy as np
import matplotlib.pyplot as plt

#1. Експоненційний розподіл
#1.1

from scipy.stats import expon
from functools import partial

f = plt.gcf()
f.set_size_inches(10, 5)
ex1 = expon(0, 1)
ex2 = expon(0, 1)
vel = ex1.rvs(10000)
vel1 = ex2.rvs(10000)
eta = vel / (vel + vel1)
h = partial(plt.hist, density=True, rwidth=0.5, bins=30)
h(eta)
plt.plot((0, 1), (1, 1), 'g')
plt.show()
```

```python
#1.2

from scipy.stats import gamma

f = plt.gcf()
f.set_size_inches(10, 5)
ex1 = expon(0, 1)
ex2 = expon(0, 1)
x = np.arange(0, 15, 0.01)
vel = ex1.rvs(10000)
vel1 = ex2.rvs(10000)
eta = vel + vel1
plt.hist(eta, bins=20, rwidth=0.5, density=True)
plt.plot(x, x * np.exp(-x), 'g')
plt.show()
```

```python
#2. Нормальний розподіл
#2.1

from scipy.stats import norm
from scipy. stats import uniform

n = norm(0,1)
n1 = norm(0,1)
theta = uniform(0, 2 * np.pi)
x = np.arange(-3, 3, 0.01)
vel = n.rvs(10000)
vel1 = n1.rvs(10000)
vel2 = theta.rvs(10000)

dzeta = vel * np.cos(vel2) + vel1 * np.sin(vel2)
plt.hist(dzeta, bins=20, rwidth=0.6, density=True)
plt.plot(x, n.pdf(x), 'g')
plt.show()
```

```python
#2.2

n = norm(0, 1)
n1 = norm(0, 1)
vel = n.rvs(10000)
vel1 = n1.rvs(10000)
vel2 = (vel**2 + vel1**2)**0.5
vel3 = np.arccos(vel/vel2)

x = np.arange(0, 4, 0.1)
y = x*np.exp((-x**2)/2)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].hist(vel2,bins=20, density=True, rwidth=0.6)
axs[0].plot(x,y,'g')
axs[1].hist(vel3,bins=20, density=True, rwidth=0.6)
#axs[1].hlines(1/3, 0, 3, 'g')
plt.show()
```

```python
#3. Рівномірний розподіл

ki = uniform(-1,2)

x = ki.rvs(100000)
x1 = ki.rvs(100000)

eta = x+ x1

plt.hist(eta,bins=20, density=True, rwidth=0.6)
plt.plot((-2, 0, 2), (0, 0.5, 0), 'g')
plt.show()
```
