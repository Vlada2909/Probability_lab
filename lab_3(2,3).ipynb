```python
# second task
import numpy as np
import itertools

def task_2a(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    return len(b)/(len(a))



def empiric(n, k, p):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 100000

def task_2b(n, k, p,q):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0 or a[i][j] % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    return len(b)/(len(a))


def empiric_b(n, k, p,q):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0 or i % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 10**5

def task_2c(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                b.append(a[i][j])
                break
    return len(b)/(len(a))


def empiric_c(n, k, p):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        for i in a:
            if i % p == 0:
                b.append(i)
                break

    return len(b) / 10**5


val = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("------------------------Second task (a)----------------------------")
for i in val:
    print("Theor probability n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_2a(i[0], i[1], i[2])))
    print("Empiric probability is {}".format('%.30f' % empiric(i[0], i[1], i[2])))

print("------------------------Second task (b)----------------------------")
for i in val:
    print("Theor prob n={}, k={}, p={}, q={} is {}".format(i[0], i[1], i[2], i[3], '%.30f' % task_2b(i[0], i[1], i[2], i[3])))
    print("Empiric probab {}".format('%.30f' % empiric_b(i[0], i[1], i[2], i[3])))

print("-------------------------Second task (c)----------------------------")
for i in val:
    print("Theor prob n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_2c(i[0], i[1], i[2])))
    print("Empiric prob is {}".format('%.30f' % empiric_c(i[0], i[1], i[2])))
```

```python
#third task
import itertools
import numpy as np

def condition(comb):
    a = 0
    a1 = 0
    a2 = 0
    l = len(comb)
    for i in comb:
        if len(i) == len(set(i)):
            a += 1
        elif len(i)  == len(set(i)) + 1:
            a1 += 1
        elif len(i) == len(set(i)) + 2:
            a2 += 1
    return a / l, a1 / l, a2 / l


val = [[8,4],[13,5]]
for i in val:
    pairs = [x for x in range(1, i[0] + 1)] * 2
    comb = list(itertools.combinations(pairs, r=2 * i[1]))
    f, g, h = condition(comb)
     
    n = 100000
    b = []
    for j in range(n):
        a = np.random.choice(pairs, size=2 * i[1], replace=False)
        b.append(a)
    f1, g1,h1  = condition(b)
     
    print(i)
    print("Theoretical prob that there is no paired boots:", f)
    print("Empirical prob that there is no paired boots:", f1)
    print("Theoretical prob that there is only one pair:", g)
    print("Empirical prob that there is no paired boots:", g1)
    print("Theoretical prob that there are two pairs:", h)
    print("Empirical prob that there is no paired boots:", h1)
```
